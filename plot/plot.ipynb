```python
import sys

import matplotlib
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

import numpy as np
from numpy.random import randn
from scipy import array, newaxis
```

```python
#figure 1: méthode implicite call
plot = np.loadtxt(r"D:\Projet-C\result\call_implicite.txt")
x_call_i = plot[:,0]
y_call_i = plot[:,1]
z_call_i = plot[:,2]
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_trisurf(x_call_i, y_call_i, z_call_i, cmap=cm.jet, linewidth=0)
ax.view_init(20, 230)
fig.tight_layout()
plt.title("Surface de prix d'une option call (méthode implicite)")
plt.xlabel("S")
plt.ylabel("T - t")
ax.set_zlabel('C(S,t)')
plt.show() 
```
